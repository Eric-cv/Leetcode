Given a string, find the length of the longest substring without repeating characters.

Example 1:

Input: "abcabcbb"
Output: 3 
Explanation: The answer is "abc", with the length of 3. 
Example 2:

Input: "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

solution1(from csdn):
思路：从左往右依次遍历字符串s，如果不在substr中，则加入，如果item在substr中，首先判断目前的子字符串的长度是否为当前最长，然后我们思考下一个子字符串应该是从重复的这个字符的下一个开始的，因为如果在这个字符之前那么必然还是会遇见重复的item的。全部遍历后 判断当前长度是否为最长。

```python
# solution1
def lengthOfLongestSubString(s):
    '''
    :in type  : str
    :out type : int 
    '''
    # 输入s为空时,返回长度为0
    if not s:
        return 0
    # 非空子字符串的长度最小为1
    longestlength = 1
    substr = ''  # 存放无重复子字符串 
    for item in s:  # 遍历s
        if item not in substr:
            substr += item  # 不是重复的字符就存到sub中
        else: # 遇到重复的字符
            if len(substr)>longestlength:
                longestlength = len(substr) # 更新当前字符串长度
            # 从子字符串中存储的重复字符的下一个字符开始切片，作为新的子字符串
            # 继续执行循环过程
            substr += item
            substr = substr[substr.index(item)+1:]

    # 总的更新一次最大字符长度
    if len(substr)>longestlength:
        longestlength = len(substr)
    return longestlength

s = ['a','b','c','a','b','c','b','b']
res = lengthOfLongestSubString(s)
print(res)
```
